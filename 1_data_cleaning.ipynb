```python
import pandas as pd
```

```python
transactions = pd.read_csv('Retail_Data_Transactions.csv')
response = pd.read_csv('Retail_Data_Response.csv')

transactions.head(), response.head()
```

```python
transactions.isnull().sum()
response.isnull().sum()
```

```python
transactions = transactions.drop_duplicates()
response = response.drop_duplicates()

transactions = transactions.dropna()
response = response.dropna()
```

```python
merged_data = pd.merge(transactions, response, on='customer_id', how='inner')
merged_data.head()
```

```python
merged_data['trans_date']= pd.to_datetime(merged_data['trans_date'])
merged_data['response']= merged_data['response'].astype('int64')
merged_data
```

```python
merged_data.to_csv('Cleaned_Retail_Data.csv', index=False)
print('cleaned dataset saved successfully')
```

```python
import pandas as pd

# Load your cleaned data again
data = pd.read_csv("Cleaned_Retail_Data.csv")

# Ensure trans_date is in datetime format
data['trans_date'] = pd.to_datetime(data['trans_date'], errors='coerce')

# Extract Month and Year correctly
data['Month'] = data['trans_date'].dt.month
data['Year'] = data['trans_date'].dt.year

# Verify
print(data[['trans_date', 'Month', 'Year']])
```

```python
monthly_sales = data.groupby('Month')['tran_amount'].sum().reset_index()
monthly_sales.columns = ['Month', 'Total_Sales']
print(monthly_sales)
```

```python
data.to_csv("Cleaned_Retail_Data.csv", index=False)
```
