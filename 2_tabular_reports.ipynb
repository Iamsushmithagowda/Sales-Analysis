```python
import pandas as pd

# Load your cleaned data
data = pd.read_csv("Cleaned_Retail_Data.csv")

# Ensure 'trans_date' is recognized as a datetime
data['trans_date'] = pd.to_datetime(data['trans_date'], errors='coerce')

# Extract Month and Year correctly
data['Month'] = data['trans_date'].dt.month
data['Year'] = data['trans_date'].dt.year

# Check if months are extracted correctly
print(data[['trans_date', 'Month']].head(5))
```

```python
monthly_sales = data.groupby('Month')['tran_amount'].sum().reset_index()
monthly_sales.columns = ['Month', 'Total_Sales']
print(monthly_sales)
```

```python
yearly_sales = data.groupby('Year')['tran_amount'].sum().reset_index()
yearly_sales.columns = ['Year', 'Total_Sales']
print(yearly_sales)
```

```python
response_sales = data.groupby('response')['tran_amount'].sum().reset_index()
response_sales.columns = ['Response', 'Total_Sales']
print(response_sales)
```

```python
with pd.ExcelWriter('Report/Summary_Reports.xlsx') as writer:
    monthly_sales.to_excel(writer, sheet_name='Monthly Sales', index=False)
    yearly_sales.to_excel(writer, sheet_name='Yearly Sales', index=False)
    response_sales.to_excel(writer, sheet_name='Response Sales', index=False)

print("✅ Summary reports successfully saved to 'Report/Summary_Reports.xlsx'")
```
